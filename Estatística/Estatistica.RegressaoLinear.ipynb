```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
dados = [
    [1, 135, 12],
    [2, 115, 14],
    [3, 146, 16],
    [4, 167, 19],
    [5, 165, 22],
    [6, 176, 24],
    [7, 98, 7],
    [8, 136, 17],
    [9, 163, 18],
    [10, 119, 11]
]

# Converter para um array NumPy
dados_array = np.array(dados)

# Separar as colunas
numero_vp = dados_array[:, 0]  # Número do Vice-Presidente
Salário = dados_array[:, 1]  # Salário
Bonificações = dados_array[:, 2]  # Bonificações

# Exibir os arrays separados
print("Número do VP:", numero_vp)
print("Salário:", Salário)
print("Bonificações:", Bonificações)
```

```python
plt.scatter(Salário, Bonificações)
plt.xlabel('salário')
plt.ylabel('bonificações')
plt.show()
print(" Relção linear boa, é perceptível a presença de uma dependencia linear")
```

```python
correlation = np.corrcoef(Salário, Bonificações)[0, 1]
print(f"Coeficiente de Correlação de Pearson: {correlation}")
```

```python
X = np.array([135, 115, 146, 167, 165, 176, 98, 136, 163, 119])  # Salário
Y = np.array([12, 14, 16, 19, 22, 24, 7, 17, 18, 11])  # Bonificações

regression = LinearRegression()
regression.fit(X.reshape(-1, 1), Y)

# Obter os coeficientes
slope = regression.coef_[0]
intercept = regression.intercept_

# Imprimir a equação da regressão estimada
print(f"Equação da regressão estimada: Y = {slope:.2f} * X + ({intercept:.2f})")

# Prever os valores usando o modelo ajustado
Y_pred = slope * X + intercept

# Plotar o gráfico de dispersão e a reta de regressão
plt.scatter(X, Y, label='Dados Observados')
plt.plot(X, Y_pred, color='red', label=f'Regressão: Y = {slope:.2f} * X + {intercept:.2f}')
plt.xlabel('Salário')
plt.ylabel('Bonificações')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
# criando um data frame
data = pd.DataFrame({
    'Salario': [135, 115, 146, 167, 165, 176, 98, 136, 163, 119],
    'Bonificacoes': [12, 14, 16, 19, 22, 24, 7, 17, 18, 11]
})
X = data[['Salario']]
y = data['Bonificacoes']

# Criar um modelo de regressão linear
regression = LinearRegression()

# Ajustar o modelo aos dados
regression.fit(X, y)

# Prever os valores usando o modelo
y_pred = regression.predict(X)

# Calcular a média das bonificações
mean_y = y.mean()

# Calcular SSE, SST e SSR
SSE = ((y - y_pred) ** 2).sum()
SST = ((y - mean_y) ** 2).sum()
SSR = ((y_pred - mean_y) ** 2).sum()

print(f"SSE (Soma dos Quadrados dos Erros): {SSE:.2f}")
print(f"SST (Soma Total dos Quadrados): {SST:.2f}")
print(f"SSR (Soma dos Quadrados da Regressão): {SSR:.2f}")
```

```python
# Calcular R²
R = 1 - (SSE / SST)

print(f"Coeficiente de Determinação (R²): {R:.2f}")

# Comentando a eficiência do ajuste
if R == 1:
    print("O modelo explica 100% da variação na variável dependente. Ajuste perfeito.")
elif R > 0.7:
    print("O modelo tem um bom ajuste,uma alta proporção da variação na variável dependente.")
elif R > 0.5:
    print("O modelo tem um ajuste razoável,  uma proporção moderada da variação na variável dependente.")
else:
    print("O modelo tem um ajuste fraco,  uma baixa proporção da variação na variável dependente.")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

# Dados fornecidos
preco_venda = [400000, 370000, 382500, 300000, 305000, 320000, 321000, 445000, 377500, 460000, 265000, 299000, 385000, 430000, 214900, 475000, 280000, 457000, 210000, 272500, 268000, 300000, 477000, 292000, 379000, 295000, 499000, 292000, 305000, 520000, 308000, 316000, 355500, 225000, 270000, 253000, 310000, 300000, 295000, 478000]
area_util = [2704, 2096, 2737, 1800, 1066, 1820, 2700, 2316, 2448, 3040, 1500, 1448, 2400, 2200, 1635, 2224, 1738, 3432, 1175, 1393, 1196, 1860, 3867, 1800, 2722, 2240, 2174, 1650, 2000, 3350, 1776, 1850, 2600, 1300, 1352, 1312, 1664, 1700, 1650, 2400]

# Calcula a regressão linear
slope, intercept, R, p_value, std_err = linregress(preco_venda, area_util)

# Cria uma função para a linha de regressão
def regression_line(x):
    return slope * x + intercept

# Cria o gráfico de dispersão
plt.scatter(preco_venda, area_util, label='Dados Originais')

# Adiciona a linha de regressão ao gráfico
plt.plot(preco_venda, regression_line(np.array(preco_venda)), color='red', label='Reta de Regressão')

# Adiciona o coeficiente de determinação ao gráfico
plt.text(min(preco_venda), max(area_util), f'R² = {R**2:.2f}', ha='left', va='bottom', color='blue')

# Adiciona rótulos e título
plt.xlabel('Preço Venda')
plt.ylabel('Área Útil')
plt.title('Gráfico de Dispersão com Reta de Regressão')
print(f"Equação de Regressão Linear: Área Útil = {intercept:.5f} + {slope:.5f} * Preço Venda")
print(f"Coeficiente de Determinação (R²): {R**2:.2f}")
if R == 1:
    print("O modelo explica 100% da variação na variável dependente. Ajuste perfeito.")
elif R > 0.7:
    print("O modelo tem um bom ajuste, uma alta proporção da variação na variável dependente.")
elif R > 0.5:
    print("O modelo tem um ajuste razoável, uma proporção moderada da variação na variável dependente.")
else:
    print("O modelo tem um ajuste fraco, uma baixa proporção da variação na variável dependente.")

# Exibe a legenda
plt.legend()

# Exibe o gráfico
plt.show()
```
